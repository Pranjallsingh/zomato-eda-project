## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.

# Convert numeric columns to numeric type
df['Average Cost for two'] = pd.to_numeric(df['Average Cost for two'], errors='coerce')
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
df['Price range'] = pd.to_numeric(df['Price range'], errors='coerce')

# Fill missing values with simple placeholder values
df['Cuisines'] = df['Cuisines'].fillna('Not specified')
df['Rating text'] = df['Rating text'].fillna('Not Rated')
df['Has Table booking'] = df['Has Table booking'].fillna('No')
df['Has Online delivery'] = df['Has Online delivery'].fillna('No')
df['Switch to order menu'] = df['Switch to order menu'].fillna('No')

# Create a country name column for easier storytelling
country_map = {
    1: 'India', 14: 'Australia', 30: 'Brazil', 37: 'Canada', 94: 'Indonesia',
    148: 'New Zealand', 162: 'Philippines', 166: 'Qatar', 184: 'Singapore',
    189: 'South Africa', 191: 'Sri Lanka', 208: 'Turkey', 214: 'United Arab Emirates',
    215: 'United Kingdom', 216: 'United States'
}
df['Country'] = df['Country Code'].map(country_map).fillna('Other')

# Remove duplicate rows if any
df = df.drop_duplicates().reset_index(drop=True)

print('Data wrangling completed.')
print('Missing values after cleaning:')
print(df.isnull().sum()[df.isnull().sum() > 0])
print('Shape after cleaning:', df.shape)
```

### What all manipulations have you done and insights you found?

I converted the important numeric columns into the correct numeric format, filled missing values with simple placeholders, and created a country name column to make the analysis easier to read. I also removed duplicate rows so that the charts and summary statistics reflect the true dataset. These steps help make the analysis cleaner and more reliable for business interpretation.
